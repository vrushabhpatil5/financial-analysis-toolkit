# NPV & IRR Calculator

This notebook demonstrates how Net Present Value (NPV) and Internal Rate of Return (IRR)
are calculated and used in investment decision-making.

```python
import numpy as np

# Sample cash flows: Initial investment followed by yearly cash inflows
cash_flows = [-100000, 30000, 35000, 40000, 45000]
discount_rate = 0.10

npv = np.npv(discount_rate, cash_flows)
irr = np.irr(cash_flows)

npv, irr
```
